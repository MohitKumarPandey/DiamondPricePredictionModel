```python
import numpy as np 
import pandas as pd 
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge 
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/MohitKumarPandey/DiamondPricePredictionModel/main/diamonds.csv")
df
```

```python
df['cut']=df['cut'].str.lower()
```

```python
# encoding of data 
cut_encoder=LabelEncoder()
color_encoder=LabelEncoder()
clarity_encoder=LabelEncoder()
df['cut']=cut_encoder.fit_transform(df['cut'])
df['color']=color_encoder.fit_transform(df['color'])
df['clarity']=clarity_encoder.fit_transform(df['clarity'])
```

```python
# spliting of data 
X=df.drop('price',axis=1)
y=df['price']
```

```python
# train test split 
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=42)
```

```python
# model training and prediction 
model=LinearRegression()
model.fit(x_train,y_train)
carat=float(input("enter the carat"))
cut=input("select the cut type --> [Ideal, Premium, Good, Very Good,Fair]").lower()
color=input("select the color--> [E,I,J,H,F,G,D] ").upper()
clarity=input("select the clarity --> [SI2,SI1,VS1,VS2,VVS2,VVS1,I1,IF]").upper()
depth=float(input("enter the depth% of diamond"))
table=float(input("enter the table % of diamond"))
x=float(input("enter the length of diamond (in mm"))
y=float(input("enter the width of diamond (in mm"))
z=float(input("enter the height of diamond (in mm"))

cut=cut_encoder.transform([cut])
color=color_encoder.transform([color])
clarity=clarity_encoder.transform([clarity])
new_data=pd.DataFrame({
        "carat":[carat],
        "cut":[cut[0]],
       "color":[color[0]],
      "clarity":[clarity[0]],
      "depth":[depth],
      "table":[table],
      "x":[x],
      "y":[y],
      "z":[z]
})


# model prediction 
predict_price=model.predict(new_data)
print("Estimated price :",predict_price[0])
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error

y_pred=model.predict(x_test)
r2=r2_score(y_test,y_pred)
print("r2 score ",r2)
```

```python
# mean_absolute_error
mae=mean_absolute_error(y_test,y_pred)
print("mean_absolute_error",mae)
```

```python
# mean_squared_error
mse=mean_squared_error(y_test,y_pred)
print("mean_squared_error",mse)
```

```python
# root_mean_squared_error

rmse=np.sqrt(mse)
print("root_mean_squared_error",rmse)
```
